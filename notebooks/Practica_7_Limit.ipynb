```sql
# Informacion de Venta por Vendedor y Cliente
SELECT
   vt.ID_Venta, vt.Venta,
   CONCAT(vd.Nombre, ' ', vd.Apellido) AS Vendedor, 
   c.Nombre AS Cliente
FROM ventas AS vt
JOIN vendedor AS vd
 ON vt.ID_Vendedor = vd.ID_Vendedor 
JOIN clientes AS c ON vt.ID_Cliente = c.ID_Cliente
ORDER BY vt.ID_Venta
LIMIT 30;

```

```sql
# generar vista
CREATE VIEW Venta_por_Vendedor_y_Cliente AS
SELECT
   vt.ID_Venta, vt.Venta,
   CONCAT(vd.Nombre, ' ', vd.Apellido) AS Vendedor, 
   c.Nombre AS Cliente
FROM ventas AS vt
JOIN vendedor AS vd
 ON vt.ID_Vendedor = vd.ID_Vendedor 
JOIN clientes AS c ON vt.ID_Cliente = c.ID_Cliente
ORDER BY vt.ID_Venta;
```

Vista Ventas por Vendedor y Cliente



```sql
# Informacion de venta por cliente y zona y porducto
USE base;
SELECT
   vt.ID_Venta, vt.Venta,
   c.Nombre AS Cliente,
   vt.ID_Zona,
   p.Producto
FROM ventas AS vt
JOIN clientes AS c ON vt.ID_Cliente = c.ID_Cliente
JOIN producto AS p ON vt.ID_producto = p.ID_Producto
ORDER BY vt.ID_Venta
LIMIT 30;
```

```sql
# generar vista la venta por cliente y zona y porducto
CREATE VIEW Venta_por_Cliente_y_Zona_y_Producto_or AS
SELECT
   vt.ID_Venta, vt.Venta,
   c.Nombre AS Cliente,
   vt.ID_Zona,
   p.Producto
FROM ventas AS vt
JOIN clientes AS c ON vt.ID_Cliente = c.ID_Cliente
JOIN producto AS p ON vt.ID_producto = p.ID_Producto
ORDER BY c.Nombre, vt.ID_Zona, p.Producto;
```

26\. Encontrar al vendedor que no ha vendido nada

```sql
SELECT
    vd.ID_Vendedor,
    CONCAT(vd.Nombre, ' ', vd.Apellido) AS Vendedor
FROM vendedor AS vd
LEFT JOIN ventas AS vt ON vd.ID_Vendedor = vt.ID_Vendedor
WHERE vt.ID_Venta IS NULL;

```

28\. (CROSS JOIN) Genera una vista para traer todas las combinaciones de:  
\--Nombre de productos y nombre de clientes

```sql
# Crear Vista Nombre de Productos y Nombre de Clientes
CREATE VIEW Nomnbre_de_Prod_y_NombreClientes AS
SELECT
    p.Producto,
    c.Nombre AS Nombre_Cliente
FROM producto AS p
CROSS JOIN clientes AS c;
```

\-- 1 Una vista con clientes que tengan la calificación "A" y "B"

\-- 2 Lista de vendedores que estén casados y sean mayores de 50 años con los vendedores solteros y menores de 30

```sql
USE base;
CREATE VIEW Vista_Clientes_Calificados AS
SELECT ID_Cliente, Nombre, Direccion, Pais, Telefono, Clasificacion_credito
FROM clientes
WHERE Clasificacion_credito IN ('A', 'B');
```

```sql
CREATE VIEW Vista_Vendedores_Especiales_d AS
SELECT ID_Vendedor, Nombre, Apellido, Estado_Civil, Edad
FROM vendedor
WHERE (Estado_Civil = 'Casado' AND Edad > 50)
UNION
SELECT ID_Vendedor, Nombre, Apellido, Estado_Civil, Edad
FROM vendedor
WHERE (Estado_Civil = 'Soltero' AND Edad < 30);
```
