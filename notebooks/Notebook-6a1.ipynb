### Crear las Tablas

Voy a crear dos tablas, una para los usuarios (`Users`) y otra para los pedidos (`Orders`).

```sql
CREATE TABLE Users (UserID INT, UserName VARCHAR(50),
    PRIMARY KEY (UserID));
```

```sql
CREATE TABLE Orders (
    UserID INT,
    OrderID INT,
    PRIMARY KEY (OrderID));
```

### Insertar Datos en las Tablas

```sql
INSERT INTO Users (UserID, UserName) VALUES (123, 'Bob');
INSERT INTO Users (UserID, UserName) VALUES (124, 'Alice');
INSERT INTO Users (UserID, UserName) VALUES (125, 'Carrie');
```

```sql
INSERT INTO Orders (UserID, OrderID) VALUES (123, 333);
INSERT INTO Orders (UserID, OrderID) VALUES (124, 111);
INSERT INTO Orders (UserID, OrderID) VALUES (126, 666);
```

### Comprobar los Datos

```sql
SELECT * FROM Users;
```

```sql
SELECT * FROM Orders;
```

```sql
SELECT Users.UserID, Users.UserName, Orders.OrderID
FROM Users
INNER JOIN Orders ON Users.UserID = Orders.UserID;
```

```sql
SELECT Users.UserID, Users.UserName, Orders.OrderID
FROM Users
LEFT JOIN Orders ON Users.UserID = Orders.UserID;
```

```sql
SELECT Users.UserID, Users.UserName, Orders.OrderID
FROM Users
RIGHT JOIN Orders ON Users.UserID = Orders.UserID;
```

```sql
-- Parte LEFT JOIN
SELECT Users.UserID, Users.UserName, Orders.OrderID
FROM Users
LEFT JOIN Orders ON Users.UserID = Orders.UserID

UNION

-- Parte RIGHT JOIN
SELECT Users.UserID, Users.UserName, Orders.OrderID
FROM Users
RIGHT JOIN Orders ON Users.UserID = Orders.UserID;
```
