### Paso 1: Crear el Schema

```sql
CREATE DATABASE practica_6a;
USE practica_6a;
```

### Paso 2: Crear las Tablas

Voy a crear dos tablas, una para los usuarios (`Users`) y otra para los pedidos (`Orders`). Voy a definir las llaves primarias en este paso.

```sql
USE practica_6a;
CREATE TABLE Users 
(UserID INT, UserName VARCHAR(50),
PRIMARY KEY (UserID));
```

```sql
CREATE TABLE Orders 
( UserID INT, OrderID INT,
PRIMARY KEY (OrderID),
FOREIGN KEY (UserID) REFERENCES Users(UserID));
```
